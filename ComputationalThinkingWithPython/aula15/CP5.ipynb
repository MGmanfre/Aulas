```python
import pandas as pd
 
df = pd.DataFrame({
    'produto': ['A','B','C','D'],
    'preco': [10.0,12.5,9.0, 13.0],
    'desconto(%)': [10 ,None ,0 ,20 ],
    })
 
 
def normPreco():
    X = df['preco']
    Xmin = df['preco'].min()
    Xmax = df['preco'].max()    
    df['normPreco'] = X.apply(lambda X: (X - Xmin) / (Xmax - Xmin))
    return df['normPreco']
 
print(normPreco())
```

```python
import pandas as pd

df = pd.DataFrame({
    "nota": [5.9, 6.0, 7.5, 9.0, 5.95],
    "frequencia": [8.8, 0.75, 0.7, 0.95, 8.9]
})
```

```python
def status(x):
    if x['nota'] >=6 and x['frequencia'] >=0.75:
        return "Aprovado"
    elif x['nota'] >= 6 and x['frequencia'] < 0.75:
        return "Recuperação"
    elif x['nota'] < 6 and x['frequencia'] >= 0.75:
        return "Recuperação"
    elif x['nota'] < 6 and x['frequencia'] < 0.75:
        return "Reprovado"
    
df['status'] = df.apply(status, axis=1)
df
```

```python
import pandas as pd
df = pd.DataFrame({
    "id": [1,2,3,4,5],
    "notal": [6.0, 5.0, 8.0, 9.5, 7.0], 
    "nota2": [7.0, 6.0, 7.0, 8.0, 7.0],
    "faltas": [1, 3, 0, 2, 5,],
    "turma": ["A", "B", "A", "B", "A"]
})

df
```

```python
import numpy as np


def calcularMedia(x):
    media = np.mean(x[['notal', 'nota2']])
    return media

df['media'] = df.apply(calcularMedia, axis=1)
df['status'] = df['media'].apply(lambda x: 'Aprovado' if x >= 7 else 'Reprovado')

df
```
