```python
df = pd.DataFrame({
    'A': [1,2,3,4,5],
    'B': [10,20,30,40,50],
    'C': [100,200,300,400,500],
})

df['A_dobro'] = df['A'].apply(lambda z: z*2)
df
```

```python
df['soma_coluna'] = df.apply(lambda x: x['A'] + x['B'] + x['C'], axis=1)
df
```

```python
import numpy as np

def calcularMedia(x):
    media = np.mean([x['A'],x['B'],x['C']])
    return media

df['media_coluna'] = df.apply(calcularMedia, axis=1)
df
```

```python
df['categoria'] = df['A'].apply(lambda x: 'Alto' if x>3 else 'Baixo')
df
```

```python
df[['A_norm', 'B_norm', 'C_norm']] = (df[['A', 'B', 'C']] - df[['A', 'B', 'C']].min()) / (df[['A', 'B', 'C']].max() - df[['A', 'B', 'C']].min())
df
```
